```python
# 建立关键词索引

with open("faiss_data/stop_words.txt", "r", encoding="utf-8") as f:
    stop_words = f.read().splitlines()
print(stop_words)

```

```python
with open("faiss_data/input.txt", "r", encoding="utf-8") as f:
    input_data = f.read().splitlines()
print(input_data)
```

```python
keywords = {}
for i, line in enumerate(input_data):
    keywords[line] = keywords.get(line,[]) + [i]
    processed_keywords = jieba.lcut(line)
    processed_keywords = [word for word in processed_keywords if word not in stop_words]
    for word in processed_keywords:
        keywords[word] = keywords.get(word, []) + [i]

with open("faiss_data/my_keywords.json", "w", encoding="utf-8") as f:
    json.dump(keywords, f, ensure_ascii=False, indent=4)

















```
